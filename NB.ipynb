```python
import cv2

# Replace "image_path.jpg" with the actual path of your image file
path=r"C:\Users\Slmss\OneDrive\Desktop\A.I\Salems IT\Marketing\Logo\Brand Logo Salem IT.png"
image = cv2.imread(path)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
threshold1 = 100
threshold2 = 200
edges = cv2.Canny(gray_image, threshold1, threshold2)
```

```python
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
# Approximate the contours to get a list of points
for contour in contours:
    epsilon = 0.02 * cv2.arcLength(contour, True)
    approx = cv2.approxPolyDP(contour, epsilon, True)

    # 'approx' will contain the list of points for each contour
    print(approx)
```
